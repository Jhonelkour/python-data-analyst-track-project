```python
import pandas as pd 

df = pd.read_csv("netflix_top10.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
date_columns = ['release_date', 'week']
for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df = df.dropna(subset=['show_title', 'runtime'])
```

```python
df['content_length'] = df['runtime'].astype(str).str.extract('(\d+)').astype(float)

df[['show_title', 'runtime', 'content_length']].head()
```

```python
df.groupby('category')['content_length'].mean().round(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(
    x='content_length',
    y='weekly_rank',
    hue='category',
    data=df,
    aspect=1.5,
    scatter_kws={'alpha':0.5}
)
plt.title("Relationship between Content Runtime and Weekly Rank")
plt.xlabel("Runtime (minutes / episodes)")
plt.ylabel("Weekly Rank (1 = Top)")
plt.show()
```

```python
corr_all = df[['content_length', 'weekly_rank']].corr().iloc[0,1]
corr_by_cat = df.groupby('category')[['content_length', 'weekly_rank']].corr().iloc[0::2, -1]

print("Correlation (All):", corr_all)
print("\nCorrelation by Category:\n", corr_by_cat)
```

```python
avg_dur_by_rank = df.groupby(['weekly_rank', 'category'])['content_length'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=avg_dur_by_rank, x='weekly_rank', y='content_length', hue='category', marker='o')
plt.title("Average Runtime vs Weekly Rank")
plt.xlabel("Weekly Rank (1 = Top)")
plt.ylabel("Average Runtime (minutes / episodes)")
plt.show()
```

```python
df['year_month'] = pd.to_datetime(df['week']).dt.to_period('M').astype(str)

avg_runtime_over_time = df.groupby(['year_month', 'category'])['content_length'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=avg_runtime_over_time, x='year_month', y='content_length', hue='category')
plt.xticks(rotation=45)
plt.title("Average Runtime Over Time (Monthly)")
plt.xlabel("Month")
plt.ylabel("Average Runtime")
plt.show()
```

# Overall Finding

There‚Äôs no strong relationship between how long a movie or show is and how well it performs in the weekly Top 10 rankings.
In other words:

Content length (runtime or number of episodes) does not determine a title‚Äôs popularity on Netflix.

This suggests that other factors ‚Äî like storytelling quality, star power, marketing, or social media buzz ‚Äî are much more influential.

## Films 

English films show a tiny negative correlation (‚àí0.06):

Slightly longer films tend to perform better.
This might reflect that successful blockbuster-style movies often have longer runtimes (e.g. action or drama hits).

Non-English films, interestingly, have a positive correlation (+0.20):

Shorter non-English films perform slightly better.
That could be cultural ‚Äî maybe audiences prefer concise storytelling in international films.

## TV Shows
Both English and Non-English series have very weak correlations, around ‚àí0.05 to ‚àí0.07.

The number of episodes doesn‚Äôt have a real impact on ranking.
It‚Äôs more about whether the show goes viral, not how long it runs.

# Conclusion

After analyzing Netflix‚Äôs weekly Top 10 dataset (2021‚Äì2023), we found no significant relationship between content runtime and its ranking performance. Correlation values ranged between ‚àí0.07 and +0.20 across categories, indicating that factors other than duration‚Äîsuch as narrative strength, cast popularity, and cultural appeal‚Äîdrive audience engagement. These insights suggest that Netflix‚Äôs content success relies more on creative impact than on runtime optimization
